```python
import polars as pl

inicial = (
    pl.scan_parquet(r"C:\Users\diogo.durao\Documents\covid_sp.parquet")
    .with_columns(
        pl.col("paciente_idade").cast(pl.Int64, strict=False),
        pl.col("paciente_dataNascimento").str.to_date("%Y-%m-%d", strict=False),
        pl.col("vacina_dataAplicacao").str.to_date("%Y-%m-%d", strict=False)
    )
)

display(inicial.collect().head())
```

```python
# Filtro Simples: Apenas pacientes do Acre
q_ac = inicial.filter(
    pl.col("paciente_endereco_uf") == "AC"
)

# Filtro Composto: Acre e idade maior que 60
q_ac_idosos = inicial.filter(
    (pl.col("paciente_endereco_uf") == "AC") &
    (pl.col("paciente_idade") > 60)
)

# Filtro com Lista: Apenas AC e AP
q_estados = inicial.filter(
    pl.col("paciente_endereco_uf").is_in(["AC", "AP"])
)

display(q_ac.collect())
display(q_ac_idosos.collect())
display(q_estados.collect())
```

```python
# Selecionar apenas colunas importantes
q_simples = inicial.select([
    "paciente_endereco_uf",
    "paciente_endereco_nmMunicipio",
    "paciente_idade",
    "vacina_nome",
    "vacina_dataAplicacao"
])

# Remover colunas menos úteis/pesadas
q_sem_codigos = inicial.drop([
    "document_id",
    "paciente_id"
])

# Seleciona tudo EXCETO as colunas listadas
q_sem_codigos = inicial.select(
    pl.exclude([
        "document_id",
        "paciente_id"
    ])
)

display(q_simples.collect())
display(q_sem_codigos.collect())
```

```python
# Categorizar faixa etária
q_categoria = inicial.with_columns(
    pl.when(pl.col("paciente_idade") >= 60)
    .then(pl.lit("Idoso"))
    .when(pl.col("paciente_idade") >= 18)
    .then(pl.lit("Adulto"))
    .otherwise(pl.lit("Menor de Idade"))
    .alias("FAIXA_ETARIA")
)

display(
    q_categoria.select([
        "paciente_idade",
        "FAIXA_ETARIA"
    ]).collect()
)
```

```python
# Análise por faixa etária
q_analise_categoria = (
    q_categoria
    .group_by("FAIXA_ETARIA")
    .agg([
        pl.len().alias("total_pessoas"),
        pl.col("paciente_idade").mean().alias("media_idade"),
        pl.col("paciente_idade").min().alias("menor_idade"),
        pl.col("paciente_idade").max().alias("maior_idade")
    ])
    .sort("media_idade", descending=True)
)

display(q_analise_categoria.collect())
```

```python
# Pegar as primeiras 10 linhas
q_head = inicial.head(10)

# Pegar da linha 100 até a 150
q_meio = inicial.slice(100, 50)

display(q_head.collect())
display(q_meio.collect())
```

```python
# Recorte de linhas específicas
q_recorte_indices = (
    inicial
    .with_row_index("id_linha")
    .filter(pl.col("id_linha").is_in([0, 10, 50]))
    .select([
        "paciente_endereco_uf",
        "paciente_endereco_nmMunicipio"
    ])
)

display(q_recorte_indices.collect())
```

```python
q_stats_mun = (
    inicial.group_by("paciente_endereco_nmMunicipio")
    .agg([
        pl.len().alias("total_vacinados"),
        pl.col("paciente_idade").mean().alias("media_idade"),
        pl.col("paciente_idade").max().alias("maior_idade"),
        pl.col("paciente_endereco_uf").first().alias("uf_origem")
    ])
    .sort("total_vacinados", descending=True)
)

display(q_stats_mun.collect())
```

```python
import polars as pl
import time

print("=== TESTE 1: Processamento SEM StringCache ===")
start_1 = time.time()

q_sem_cache = (
    pl.scan_parquet(r"C:\Users\diogo.durao\Documents\covid_sp.parquet")
    .with_columns(
        pl.col("paciente_idade").cast(pl.Int64, strict=False),
        pl.col("paciente_dataNascimento").str.to_date("%Y-%m-%d", strict=False),
        pl.col("vacina_dataAplicacao").str.to_date("%Y-%m-%d", strict=False)
    )
    .group_by(["paciente_endereco_uf", "paciente_endereco_nmMunicipio"])
    .agg([
        pl.col("paciente_idade").mean().alias("media_idade"),
        pl.col("paciente_idade").std().alias("desvio_idade")
    ])
    .sort("media_idade", descending=True)
)

df_1 = q_sem_cache.collect()
end_1 = time.time()
tempo_sem = end_1 - start_1

print(f"Tempo SEM Cache: {tempo_sem:.4f} segundos")
print(f"Uso de memória estimado (coluna UF): {df_1['paciente_endereco_uf'].estimated_size()} bytes\n")

display(df_1)
```

```python
print("=== TESTE 2: Processamento COM StringCache ===")

with pl.StringCache():
    start_2 = time.time()

    q_com_cache = (
        pl.scan_parquet(r"C:\Users\diogo.durao\Documents\covid_sp.parquet")
        .with_columns(
            pl.col("paciente_idade").cast(pl.Int64, strict=False),
            pl.col("paciente_dataNascimento").str.to_date("%Y-%m-%d", strict=False),
            pl.col("vacina_dataAplicacao").str.to_date("%Y-%m-%d", strict=False)
        )
        .group_by([
            "paciente_endereco_uf",
            "paciente_endereco_nmMunicipio"
        ])
        .agg([
            pl.col("paciente_idade").mean().alias("media_idade"),
            pl.col("paciente_idade").std().alias("desvio_idade")
        ])
        .sort("media_idade", descending=True)
    )

    df_2 = q_com_cache.collect()
    end_2 = time.time()

tempo_com = end_2 - start_2

print(f"Tempo COM Cache: {tempo_com:.4f} segundos")
print(f"Uso de memória estimado (coluna UF): {df_2['paciente_endereco_uf'].estimated_size()} bytes\n")
```
